```python
df = pd.DataFrame ([[170, 60],
			[164, 50],
			[180, 70]], index=['A','B','C'], columns = ["height","weight"])
import random
import names
for i in range(100):
    new = pd.DataFrame({'height': 160+random.randint(0,20), 'weight': 50+random.randint(0,20)}, index= [names.get_first_name()])
    df = pd.concat([df, new]) # 由於 append 將在未來的版本被廢除，因此此處使用concat的方式來新增資料。
print(df)
df = df.drop(["A","B","C"])
print(df)
```

```python
print(df)
df = df.drop(df.index[0:2])
print(df)
```
