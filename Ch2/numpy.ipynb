```python
import numpy as np
```

```python
a = np.array([2,5,4,3])
b = np.array((3,6,5,4))
print(a,b)
print(type(a), type(b))
```

```python
a = np.zeros([2, 5])
print("a:",a)
b = np.ones([2, 5])
print("b:",b)
```

```python
a = np.array([[1, 2, 4],[3, 2, 5],[1, 5, 6]])
print("Size: ", a.size)
print("Dimension: ", a.ndim)
```

```python
a = np.arange(1,17, 2)
a = a.reshape(2,4)
print(a)
```

```python
a = np.arange(1,17, 1)
print("Before Reshape: ",a)
a = a.reshape(4,4)
print("After Reshape:\n", a)
```

```python
a = np.arange(1,17, 1)
a = a.reshape(4,4)
print("a",a)
print("---")
print("a[1,2]:\n",a[1,2])
print("a[0,:]:\n",a[0,:]) 
print("a[:,0]:\n",a[:,0])
```

```python
a = np.arange(1,17, 1)
a = a.reshape(4,4)
print("Before add:\n", a)
a = a+1 # 將所有的元素都加上1
print("After add:\n",a)
a = a**2  # 將所有的元素都平方
print("After square:\n", a)
```

```python
a = np.arange(1,5, 1)
a = a.reshape(2,2)
b = np.arange(5,9, 1)
b = b.reshape(2,2)
print("sum：\n",a+b)
print("multiply: \n", a*b)
```

```python
a = np.arange(1,5, 1)
a = a.reshape(2,2)
b = np.arange(5,9, 1)
b = b.reshape(2,2)
print(np.dot(a,b))
```
