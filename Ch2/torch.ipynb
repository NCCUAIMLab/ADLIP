```python
import torch
import numpy as np
```

```python
a = torch.tensor([[1,-1],[1,-1]]) 
print(a)
b = torch.tensor(np.array([[1, 2, 3],[1,-2, -3]]))
print(b)
```

```python
a = torch.ones([2, 3], dtype=torch.float64)
print(a)
```

```python
b = torch.zeros([2, 3], dtype=torch.float64)
print(b)
```

```python
a = torch.ones([2, 3], dtype=torch.float64)
b = torch.zeros([2, 3], dtype=torch.float64) 
c = torch.stack([a,b], dim=0)
print(c)
print(c.shape)
```

```python
a = torch.ones([2, 3], dtype=torch.float64)
b = torch.zeros([2, 3], dtype=torch.float64) 
c = torch.stack([a,b], dim=1)
print(c)
print(c.shape)
```

```python
c = torch.stack([a,b], dim=2)
print(c)
print(c.shape)
```

```python
d = torch.concat([a,b])
print(d)
print(d.shape)
```

```python
x = torch.zeros(2, 1, 2, 1, 2)
print(x)
print(x.shape)
```

```python
y = torch.squeeze(x)
print(y)
print(y.shape)
```

```python
y = torch.squeeze(x, 0)
print(y.shape)
```

```python
y = torch.squeeze(x, 3)
print(y.shape)
```

```python
x = torch.tensor([1, 2, 3, 4])
y = torch.unsqueeze(x, 0)
print(y)
print(y.shape)
```

```python
y = torch.unsqueeze(x, 1)
print(y)
print(y.shape)
```
