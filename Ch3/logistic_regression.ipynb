```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt

data = sio.loadmat('train.mat')
data_test = sio.loadmat('test.mat')
x1, x2, y_train = data['x1'], data['x2'], data['y']
x1_test, x2_test, y_test = data_test['x1'], data_test['x2'], data_test['y']

X = np.concatenate((x1, x2), axis=1)
X_test = np.concatenate((x1_test, x2_test), axis=1)

lr = LogisticRegression(fit_intercept = False)
z = lr.fit(X,y_train)
theta1, theta2 = z.coef_[0]

plt.plot( np.linspace(2,9,8),-theta1/ theta2 * np.linspace(2,9,8))
plt.scatter(x1,x2,s=2,c=y_train)                                                       
plt.show()
print('theta1, theta2= ',theta1, theta2)
y_pred_test = z.predict(X_test)
print('Error rate =',len([i for i in zip(y_test.tolist(),y_pred_test) if i[0][0] != i[1]])/len(y_test))
```
